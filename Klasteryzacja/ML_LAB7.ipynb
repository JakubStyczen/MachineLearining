```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
 
X, y= make_blobs(n_samples=210, centers=3, n_features=2, cluster_std=0.5, shuffle=True,random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c='red', marker='x')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np


# Part 2: Tworzenie obiektu klasy KMeans z określonymi parametrami
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, tol=0.0001, random_state=42)

# Part 3: Dopasowanie danych do stworzonego obiektu
km.fit(X)

# Part 4: Wyznaczenie wartości zniekształceń
inertia = km.inertia_
print("Wartość zniekształceń (inertia):", inertia)

# Wizualizacja danych
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], s=70, c=km.labels_, cmap=plt.cm.prism)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()
```

```python
def elbow_method(X, max_k):
    distortions = []
    for k in range(1, max_k + 1):
        km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=42)
        km.fit(X)
        distortions.append(km.inertia_)
    
    # Wizualizacja wykresu wartości zniekształceń dla różnych liczby centroidów
    plt.plot(range(1, max_k + 1), distortions, marker='o')
    plt.xlabel('Liczba centroidów (k)')
    plt.ylabel('Wartość zniekształceń')
    plt.title('Metoda łokcia')
    plt.show()

# Przykładowe użycie
elbow_method(X, max_k=10)
```
